## Permutation

```python
from typing import List, Any

class Permutation:
    
    def _backtrack(self, elements: List[Any], path: List[Any], used: List[bool], result: List[List[Any]]):
        if len(path) == len(elements):
            result.append(path[:])
            return
        
        for i in range(len(elements)):
            if used[i]:
                continue
            used[i] = True
            path.append(elements[i])
            self._backtrack(elements, path, used, result)
            path.pop()
            used[i] = False
            
    def __call__(self, elements: List[Any]) -> List[List[Any]]:
        result = []
        used = [False] * len(elements)
        self._backtrack(elements, [], used, result)
        return result
```

```python
Permutation()(['a', 'b', 'c'])
```
