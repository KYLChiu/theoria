# Implementation

```python
def invert_bst(tree: BST) -> BST:
    def _invert_recursive(node: Optional[Node]) -> Optional[Node]:
        if node is None:
            return None

        # Recursively invert the left and right subtrees and assign them
        left_inverted = _invert_recursive(node.left)
        right_inverted = _invert_recursive(node.right)

        # Swap the left and right children
        node.left, node.right = right_inverted, left_inverted

        return node

    _invert_recursive(tree.root)
    return tree
```

# Tests

```python
test_cases = [
    TestCase(
        input_data={"tree": BST()},
        expected_output=BST(),
        description="Empty BST",
    ),
    TestCase(
        input_data={"tree": insert_multiple(BST(), [4, 2, 7, 1, 3, 6, 9])},
        expected_output=insert_multiple(BST(), [4, 7, 2, 9, 6, 3, 1]),
        description="""
        4
       / \
      2   7
     / \ / \
    1  3 6  9
    
    -> 
    
        4
       / \
      7   2
     / \ / \
    9  6 3  1
    """,
    ),
]
```

```python
def are_mirrored_bst(bst1: BST, bst2: BST) -> bool:
    def are_mirror_nodes(node1: Optional[Node], node2: Optional[Node]) -> bool:
        if node1 is None and node2 is None:
            return True
        if node1 is None or node2 is None:
            return False
        if node1.value != node2.value:
            return False

        # Left subtree of node1 vs right subtree of node2 and vice versa
        return are_mirror_nodes(node1.left, node2.right) and are_mirror_nodes(
            node1.right, node2.left
        )

    return are_mirror_nodes(bst1.root, bst2.root)


Validor(invert_bst).add_cases(test_cases).run(comparison=are_mirrored_bst)
```
